```python
import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset("tips")
df=tips.copy()
df.head()
```

```python
#lmplot dogru çizimi için kullanıldı
sns.lmplot(x="total_bill",y="tip",data=df)
```

```python
#dogru etrafındaki açık renkler sapmaları ifade ediyer // standart hatayı artırmaktadır
```

```python
sns.lmplot(x="total_bill",y="tip",hue="smoker",data=df)
```

```python
#mavi dogru=toplam ödenen tutarın bahşiş degişkenine göre ortalamsını verir
#turuncu dogru=toplam ödenen tutarın smoker kullanan kişilere göre bahşiş ortalamasını gösterir
```

```python
sns.lmplot(x="total_bill",y="tip",hue="smoker",col="time",data=df)
```
