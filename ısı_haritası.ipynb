```python
import seaborn as sns
df=sns.load_dataset("flights")
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df.describe().T
```

```python
df=df.pivot_table(values="passengers",columns="month",index="year")
#düzenli bir forma getirip okunmasını kolaylaştırdık
```

```python
df
```

```python
#heatmap kullanmadan önce tabloyu okunabilir hale getirdik
```

```python
sns.heatmap(df)
```

```python
#200 den 600 degerine gidildikçe yoğunluk artmaktadır
#yani yolcu sayısı artış halindedir
```

```python
#grafik yorumu= teknolojinin gelişmesiyle her geçen yıl insanlar hava ulaşımını tercih ettiklerini görüyoruz
#aynı şekilde yaz aylarında seyahat için yolculuk fazla olduğundan yoğunluk artışı olmuştur
```

```python
sns.heatmap(df,annot=True,fmt="d")
```

```python
#yolcu sayılarını kutulara yerleştirio öyle gözlem yapalım
```

```python
sns.heatmap(df,annot=True,fmt="d",linewidths=.10,linecolor="blue")
```
