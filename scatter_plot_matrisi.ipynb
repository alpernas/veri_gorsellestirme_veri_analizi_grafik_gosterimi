```python
#Scatterplot matrisi  //sayıasl değişkenler arasındaki ilişkiyi ifade eden saçılım gragiklerini oluşturu
import seaborn as sns
df=sns.load_dataset("iris")
df.head()
```

```python
sns.pairplot(df)
```

```python
#pairplot veri seti içerisindeki sayıasl değişkenelerin ilişkisini gösterir
```

```python
#NOT=saçılım grafiği görsellerinde eğer oluşmuş olan grafik toz bulutu şeklindeyse
#//veya yapısal formu yoksa iki değişken arasında ilişki olmadığı anlamına gelir
```

```python
#NOT2=oluşturulmuş olan scatterplot içerisinde gözlem noktaları incelendeğinde bunlar farklı köşelerde kümeleniyorsa
#//bu durumda çaprazlanma durumuyla gözlemlenecek olan gruplar vardır bunalrı göz önünde bulundurmamız gerekir
```

```python
sns.pairplot(df,hue="species")#boyut ekledik
```

```python
sns.pairplot(df,hue="species",markers=["o","s","d"])
```

```python
sns.pairplot(df,kind="reg")#her grafiğe dogru ekledi
```

```python
sns.pairplot(df,kind="reg",hue="species")
```
