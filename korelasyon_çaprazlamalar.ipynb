```python
import seaborn as sns
tips=sns.load_dataset("tips")
df=tips.copy()
df.head()
```

```python
sns.scatterplot(x="total_bill",y="tip",data=df)
```

```python
sns.scatterplot(x="day",y="total_bill",data=df)
```

```python
sns.scatterplot(x="total_bill",y="tip",hue="time",data=df)
```

```python
sns.scatterplot(x="total_bill",y="tip",hue="smoker",style="time",data=df)
```

```python
sns.scatterplot(x="total_bill",y="tip",hue="day",style="time",data=df)
```

```python
sns.scatterplot(x="total_bill",y="tip",hue="sex",size="size",data=df)
```
