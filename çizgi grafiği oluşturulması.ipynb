# Çizgi grafik ve Çaprazlamalar

```python
import seaborn as sns
df=sns.load_dataset("fmri")
df.head()
```

```python
sns.lineplot(x="timepoint",y="signal",data=df)
```

```python
#mavi çizgiler ortalamayı temsil etmektedir
#etrafındaki yogunluklar standart sapmayı temsil eden güven aralıklarıdır etmektedir
```

```python
sns.lineplot(x="timepoint",y="signal",hue="event",data=df)
```

```python
#mekanik olan veri setinin daha iyi yorumlamamızı saglar
```

```python
sns.lineplot(x="timepoint",y="signal",hue="event",style="event",data=df)
```

```python
sns.lineplot(x="timepoint",
             y="signal",
             hue="event",
             style="event",
             markers=True,
             dashes=False,
             data=df)
```
