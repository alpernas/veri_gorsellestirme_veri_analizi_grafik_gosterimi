```python
import seaborn as sns
import pandas as pd
iris=sns.load_dataset("iris")
df=iris.copy()
df.head()
```

```python
df.dtypes
```

```python
df.species=pd.Categorical(df.species)#category olarak degiştirdik
```

```python
df.dtypes
```

```python
df.shape
```

```python
sns.pairplot(df)#değişkenler arası ilişki gösterimi
```

```python
sns.pairplot(df,hue="species")
```

```python
sns.pairplot(df,hue="species",markers=["o","s","d"])
```

```python
sns.pairplot(df,hue="species",kind="reg")
```
